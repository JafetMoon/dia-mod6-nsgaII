## Aplicar NSGA-II al problema WFG1 con n=24 y k=3

```python
from pymoo.algorithms.moo.nsga2 import NSGA2
from pymoo.problems.many.wfg import WFG1
from pymoo.optimize import minimize
from pymoo.visualization.scatter import Scatter

problem = WFG1(n_var=24, n_obj=3)

algorithm = NSGA2(pop_size=200)

res = minimize(problem,
               algorithm,
               ('n_gen', 10000),
               seed=1,
               verbose=False)
```

## Graficar frente de Pareto encontrado


Se requirió aumentar el número de generaciones `n_gen` un número alto (se probó desde 100 a 10000) para converger al frente de Pareto conocido. También se varió el número de individuos con el mismo objetivo.

```python
plot = Scatter()
plot.add(problem.pareto_front(), plot_type="line", color="black", alpha=0.7)
plot.add(res.F, facecolor="none", edgecolor="red")
plot.show()
```

## Graficar conjunto de Pareto con Radviz

Con esta visualización se puede observar como las soluciones en el espacio objetivo se agrupan mayormente cerca de $f_1$ y $f_3$.

```python
from pymoo.visualization.radviz import Radviz
Radviz().add(res.F).show()
```
